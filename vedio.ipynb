```python
import cv2
from keras.models import load_model
from tensorflow.keras.utils import img_to_array 
from tensorflow.keras.utils import load_img
import numpy as np
# is the model that we trained you can also use already developed model for grater accuracy
model = load_model('./face_mask_detection_model.h5')
# model = load_model('replace with the model name')
img_width=200
img_height =200

# Haarcascade is an algo used to detect a face you can download harcascade file
face_cascade = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
#face_cascade = cv2.CascadeClassifier('replace eith haarcascade file')
cap= cv2.VideoCapture(0)

img_count_full=0
font= cv2.FONT_HERSHEY_SIMPLEX
org=(1,1)
class_lable=' '
fontscale=1
color=(225,0,0)
thickness=2 

while True:
    img_count_full+=1
    
    
    responce, color_img=cap.read()
    if responce ==False:
         break
    
    
    
    
    gray_img= cv2.cvtColor(color_img,cv2.COLOR_BGR2GRAY)
    faces= face_cascade.detectMultiScale(gray_img, 1.1, 6)
    
    
    img_count =0
    for(x,y,w,h) in faces:
        org=(x-10,y-10)
        img_count+=1
        color_face = color_img[y:y+h,x:x+w]
        # Create a folder called input and a subfolder inside input named faces it will contain all the images recorded by facecam
        cv2.imwrite('input_faces/faces/%d%dface.jpg'%(img_count_full,img_count),color_face)
        img= load_img('input_faces/faces/%d%dface.jpg'%(img_count_full,img_count),target_size=(img_width,img_height))
        
        img=img_to_array(img)/255
        img = np.expand_dims(img,axis=0)
        pred_prob = model.predict(img)
        pred=np.argmax(pred_prob)
        
        if pred==0:
            print("User With Mask - predict = ",pred_prob[0][0])
            class_lable="Mask"
            color=(255,0,0)
            # Create a folder inside input named with mask it will contain all the images that are having a mask
            cv2.imwrite('input_faces/with_mask/%d%dface.jpg'%(img_count_full,img_count),color_face)
            cv2.rectangle(color_img, (x,y) ,(x+w,y+h), (255,0,0), 3)
        else:
            print("User Without Mask - prob= ",pred_prob[0][1])
            class_lable="NO Mask"
            color=(255,0,0)
            # Create a folder inside input named without mask it will contain all the images that are not having a mask
            cv2.imwrite('input_faces/without_mask/%d%dface.jpg'%(img_count_full,img_count),color_face)
            cv2.rectangle(color_img, (x,y) ,(x+w,y+h), (0,0,255), 3)
        cv2.putText(color_img, class_lable ,org, font,fontscale,color,thickness, cv2.LINE_AA)
    cv2.imshow('live face mask detection',color_img)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
